```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Read data
df = pd.read_csv(r'C:\Users\benja\Downloads\loca\movies.csv') 
```

```python
df.head()
```

```python
# See if we have any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) 
    print('{} - {}%'.format(col, round(pct_missing*100))) 
```

```python
# Drop the rows of missing data
df = df.dropna() 
```

```python
# Check again
for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) 
    print('{} - {}%'.format(col, round(pct_missing*100))) 
```

```python
# Data types for the columns
df.dtypes
```

```python
# Change data type of columns
df['budget'] = df['budget'].astype('int64') 

df['gross'] = df['gross'].astype('int64') 
```

```python
df
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)  

df
```

```python
# Order data by gross
df.sort_values(by=['gross'], inplace=False, ascending=False) 
```

```python
# Scatter plot with Gross vs Budget 

plt.scatter(x = df['budget'], y=df['gross'])
plt.title('Gross Earnings vs Budget')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot Groos vs Budget using seaborn

sns.regplot(x = 'budget', y = 'gross', data=df, scatter_kws={"color": "purple"}, line_kws={"color": "yellow"}) 
plt.show()
```

```python
# Start looking at correlation

df.corr(numeric_only=True, method='pearson') 
```

```python
# Visualization

corr_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(corr_matrix, annot=True) 
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show() 
```

```python
# Look at Company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
corr_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(corr_matrix, annot=True) 
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show() 
```

```python
corr_matrix = df_numerized.corr(numeric_only=True)

corr_pairs = corr_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values() 

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)] 

high_corr
```
